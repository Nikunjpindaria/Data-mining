# Step 1: Load Data

```python
import pandas as pd
```

```python
data = pd.read_csv('OnlineRetail.csv',encoding = 'ISO-8859-1')
data
```

# Step 2: Drop Missing Invoices or Descriptions

```python
data_dropna = data.dropna(subset=["Description"])
data_dropna
```

```python
data_dropna = data.dropna(subset=["InvoiceNo"])
data_dropna
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
data_cleaned = data[~data['InvoiceNo'].str.startswith('C')]
data_cleaned
```

# Step 4: Keep Only Positive Quantities

```python
data_cleaned = data[data['Quantity'] > 0]
data_cleaned
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
data['Description'] = data['Description'].str.lower().str.strip()
data
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
data_uk = data[data['Country'] == 'United Kingdom']
data_uk
```

# Step 7: Remove Duplicates

```python
data_unique = data.drop_duplicates()
data_unique
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
data['TotalPrice'] = data['Quantity'] * data['UnitPrice']
data
```

# Step 9 : Check for Top 10 Selling Products

```python
top_products = data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top_products
```

# Step 10 : Identify Most Active Customers

```python
active_customers = data.groupby('CustomerID')['InvoiceNo'].count().sort_values(ascending=False).head(10)
active_customers
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data

data['Month'] = data['InvoiceDate'].dt.month
data['Hour'] = data['InvoiceDate'].dt.hour
data['Weekday'] = data['InvoiceDate'].dt.day_name()

data
```

```python
import matplotlib.pyplot as plt

popular_months = data['Month'].value_counts().sort_index()
popular_hours = data['Hour'].value_counts().sort_index()
print(popular_hours)

# Monthly
popular_months.plot(kind='bar', title='Popular Months')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.show()

# Hourly
popular_hours.plot(kind='bar', title='Popular Hours')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Orders')
plt.show()
```
